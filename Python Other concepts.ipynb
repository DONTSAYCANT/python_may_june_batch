```python
#program for calling simple functions (independent) ===> thread ===> parallel execution

import time

def function1():
    for i in range(100):
        print("python")
        
def function2():
    for i in range(100):
        print("java")
        
def function3():
    for i in range(100):
        print("groovy")

a = Thread(target = function1)
b = Thread(target = function2)
c = Thread(target = function3)


a.start()
time.sleep(5)
b.start()
a.join() #waiting for completion of a thread ==> join function
b.join()
c.start()
print("main thread")
```

```python
#program for calling simple functions (independent) ===> thread ===> parallel execution #passing arguments

def function1(p):
    for i in range(100):
        print(p)
        
def function2(q):
    for i in range(100):
        print(q)
        
def function3(x):
    for i in range(100):
        print(x)

a = Thread(target = function1, args = ("c",))
b = Thread(target = function2, args = ("cpp",))
c = Thread(target = function3, args = ("js",))


a.start()
b.start()
c.start()
print("main thread")
```
